```python
import pandas as pd
```

```python
df = pd.read_csv('orders.csv')
df
```

```python
print(df.head())
```

```python
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%Y-%m-%d')
df.dtypes
```

```python
df.isnull().sum()
```

```python
df['Ship Mode'] = df['Ship Mode'].fillna('Default Class')
df.isnull().sum()
```

```python
df.rename(columns={'Order Id': 'Order_Id', 'Order Date': 'Order_Date','Ship Mode': 'Ship_Mode', 'Postal Code': 'Postal_Code'
                  ,'Sub Category': 'Sub_Category', 'Product Id': 'Product_Id', 'cost price': 'cost_price', 
                   'List Price': 'List_Price', 'Discount Percent': 'Discount_Percent'}, inplace=True)
df.dtypes
```

```python
df[['Ship_Mode', 'Segment', 'Country', 'City', 'State', 'Region', 'Category', 'Sub_Category', 'Product_Id'] = df[['Ship_Mode', 'Segment', 'Country', 'City', 'State', 'Region', 'Category', 'Sub_Category', 'Product_Id']].apply(lambda x: x.str.rstrip())

```

```python
df[['Ship_Mode', 'Segment', 'Country', 'City', 'State', 'Region', 'Category', 'Sub_Category', 'Product_Id']] = df[['Ship_Mode', 'Segment', 'Country', 'City', 'State', 'Region', 'Category', 'Sub_Category', 'Product_Id']].apply(lambda x: x.str.rstrip())
```

```python
df
```

```python
df['Discount'] = df['List_Price'] * (df['Discount_Percent'] / 100)

# Calculate the Sale Price column
df['Sale_price'] = df['List_Price'] - df['Discount']

# Calculate the Profit column
df['Profit'] = df['Sale_price'] - df['cost_price']

# updated DataFrame
df
```

```python
df1 = df[['Order_Id', 'Order_Date', 'Ship_Mode', 'Segment', 'Country', 'City', 'State', 'Postal_Code', 'Region', 'Category'  ]]  # orders_details
df2 = df[['Order_Id', 'Sub_Category', 'Product_Id', 'cost_price', 'List_Price', 'Quantity', 'Discount_Percent', 'Discount', 'Sale_price', 'Profit']]  # Product_details

# Save the DataFrames to CSV files
df1.to_csv('orders_details.csv', index=False)  # Saving df1 to a CSV file without the index column
df2.to_csv('Product_details.csv', index=False) 
```

```python
import streamlit as st
import pandas as pd
import mysql.connector

# Function to fetch data from MySQL
def get_data_from_mysql():
    connection = mysql.connector.connect(
        host="localhost",
        user="root",
        password="mysql",
        database="retail_order"
    )
    
    query = "SELECT * FROM retail_order.product_details"  # Modify this query as needed
    data = pd.read_sql(query, connection)  # Use pandas to execute SQL query and load data into a DataFrame
    
    connection.close()
    return data

# Streamlit interface
def main():
    st.title("MySQL Data in Streamlit")

    # Fetch data from MySQL
    data = get_data_from_mysql()

    # Display data in Streamlit
    st.subheader("Data from MySQL Database:")
    st.dataframe(data)

if __name__ == "__main__":
    main()
```
